```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy import optimize
import scipy
import seaborn as sns
```

enquanto estava fazendo um estudo dos pdfs, acabou que no [último jupyter notebook](aprendizado.ipynb) no início fiz um fit do crystal ball para tentar fazer o mesmo plot disponível no pdf da aula, então vamos reaproveitar a função de lá, porém, vetorizada.

```python
def crystal_ball(x, beta, m):
  result = np.zeros_like(x, dtype=float)
  mask = x >= -beta
  result[mask] = np.exp(-x[mask]**2 / 2)
  n = (m / abs(beta))**m * np.exp(-beta**2 / 2)
  result[~mask] = n * (m / abs(beta) - beta - x[~mask])**(-m)
  return result
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y = crystal_ball(x, 0.5, 20)
```

```python
sns.lineplot(x=x, y=y)
```

```python
optimize.curve_fit(crystal_ball, x, y)
```
