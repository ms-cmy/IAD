```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math
import ROOT
```

```python
def my_func(x, par):
    return math.sin(x[0] * 2) / x[0]
```

```python
myobj = ROOT.TF1("myfunc", my_func, 0.1, 10, 0)
c = ROOT.TCanvas("c", "Canvas", 800, 600)
myobj.Draw()
c.Draw()
```
